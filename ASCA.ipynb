```python
# prompt: write a code to get an audio signal in WAV format and plot the mel-frequency of that

import numpy as np
import librosa
import matplotlib.pyplot as plt

# Load the audio signal from the specified path
x, sr = librosa.load('/content/datasets/638368586477172106682/words/word_12/keystroke_2.wav')

# Compute the mel-frequency spectrogram
mel_spec = librosa.feature.melspectrogram(y=x, sr=sr)

# Convert the mel-spectrogram to decibels
log_mel_spec = librosa.power_to_db(mel_spec, ref=np.max)

# Plot the mel-frequency spectrogram
plt.figure(figsize=(10, 4))
librosa.display.specshow(log_mel_spec, x_axis='time', y_axis='mel', sr=sr)
plt.title('Mel-frequency spectrogram')
plt.colorbar(format='%+02.0f dB')
plt.show()
```

```python
# prompt: write a code to get an audio signal in WAV format and plot the spectrogram of that using STFT

import numpy as np
import librosa
import matplotlib.pyplot as plt

# Load the audio signal from the specified path
x, sr = librosa.load('/content/datasets/638368586477172106682/words/word_12/keystroke_2.wav')

# Compute the short-time Fourier transform (STFT) of the audio signal
stft = librosa.core.stft(x, n_fft=2048, hop_length=512)

# Compute the magnitude of the STFT
magnitude = np.abs(stft)

# Compute the logarithm of the magnitude of the STFT
log_magnitude = librosa.power_to_db(magnitude, ref=np.max)

# Plot the spectrogram of the audio signal
plt.figure(figsize=(10, 4))
librosa.display.specshow(log_magnitude, x_axis='time', y_axis='log', sr=sr)
plt.title('Spectrogram of the audio signal')
plt.colorbar(format='%+02.0f dB')
plt.show()
```
